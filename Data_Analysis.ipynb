```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fert = pd.read_csv('gapminder_total_fertility.csv',',')
life = pd.read_excel('gapminder_lifeexpectancy.xlsx')
pop = pd.read_excel('gapminder_population.xlsx')
#fert.head()
```

```python
fert.shape,life.shape,pop.shape
```

```python
fert_df = fert.set_index('Total fertility rate')
life_df = life.set_index('Life expectancy')
pop_df = pop.set_index('Total population')
fert_df.head()
```

```python
life_df.columns=life_df.columns.astype(int)
life_df.columns
```

```python
pop_df.columns = pop_df.columns.astype(int)
pop_df.columns
```

```python
fert_df.columns = fert_df.columns.astype(int)
fert_df.columns
```

```python
fert_df.index.name='Country'
life_df.index.name='Country'
pop_df.index.name='Country'
```

```python
fert_df = fert_df.reset_index()
life_df = life_df.reset_index()
pop_df = pop_df.reset_index()
```

```python
fert_df = fert_df.melt(id_vars='Country', var_name='year', value_name='Fertility rate')
fert_df
```

```python
life_df = life_df.melt(id_vars='Country', var_name ='year', value_name = 'Life expectancy')
life_df
```

```python
pop_df = pop_df.melt(id_vars='Country', var_name ='year', value_name = 'Total Population')
pop_df
```

```python
df=fert_df.merge(pop_df)
df
```

```python
df1=df.merge(life_df)
df1
```

```python
continent = pd.read_csv('continents-Copy1.csv',';')
#df = df.merge(continents)
continent = continent.rename(columns={'country':'Country'})
df2 = df1.merge(continent)
df2
```

```python
df_all_subset = df2.loc[df2['year'].isin(range(1920,2016))]
x_min = df_all_subset['Life expectancy'].min()
x_max = df_all_subset['Life expectancy'].max()
y_min = df_all_subset['Fertility rate'].min()
y_max = df_all_subset['Fertility rate'].max()
z_min = df_all_subset['Total Population'].min()
z_max = df_all_subset['Total Population'].max()


for year in range(1960,1961):
    df_subset = df2.loc[df['year'].isin([year])]
    if not df_subset.empty:
        X,Y = df_subset['Life expectancy'],df_subset['Fertility rate']
        yp = df_subset['Total Population']
        Z = (yp/z_max)
        my_cmap = plt.get_cmap('tab20')
        fig = plt.figure(figsize=(12,8))
        plt.scatter(X,Y,Z*1000,Z,cmap=my_cmap)
        plt.title(f'correlation between Life expectancy and Fertility rate {year} ')
        plt.xlabel('Life expectancy')
        plt.ylabel('Fertility rate')
        plt.axis((x_min, x_max, y_min, y_max))
        fig.savefig(f'lifeexpe_{year}.png')
        plt.close()
```

```python
import imageio

images = []

for i in range(1960, 2016):
    filename = f'lifeexpe_{i}.png'
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=5)
```
